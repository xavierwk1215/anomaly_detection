```python
import torch
import numpy as np
import cv2
import requests
from io import BytesIO
from torchvision.transforms import functional as F
```

```python
# 파일 경로
checkpoint_path = '/content/drive/MyDrive/ds_study/Final_team_3/Code/taebin3/Learning-to-Adapt-to-Unseen-Abnormal-Activities/pretrain/20iters_exp0_seed1_lr0.001_split4.pkl'
```

```python
# 모델 로드
model = YourModel()  # 모델 아키텍처에 대한 클래스 생성자를 사용하세요
checkpoint = torch.load(checkpoint_path)
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()
```

```python
# 동영상 URL
video_url = 'YOUR_VIDEO_URL_HERE'

# 동영상 다운로드 및 프레임 추출
response = requests.get(video_url)
video_data = BytesIO(response.content)
cap = cv2.VideoCapture(video_data)
```

```python
# 예측 생성
predictions = []
with torch.no_grad():
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        # 프레임을 모델 입력 형식으로 변환
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = F.to_tensor(frame).unsqueeze(0)
        outputs = model(frame)
        prediction = outputs.argmax(dim=1).item()
        predictions.append(prediction)
```

```python
# 예측 결과 확인
print(predictions)

# 작업이 끝난 후 동영상 캡처 해제
cap.release()
```
